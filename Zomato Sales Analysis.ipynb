```python
def HandleRate(value):
    value = str(value).split('/')[0]  
    
   
    if value.replace('.', '', 1).isdigit():  
        return float(value)
    else:
        return None  

data['rate'] = data['rate'].apply(HandleRate)
print(data.head())
```

```python
data.info()
```

# Task.1. What type of restaurant do the majority of customers order from?

```python
sns.countplot(x=data['listed_in(type)'])
plt.xlabel('rest_type')
```

# Task.2. How many votes has each type of restaurant received from customers?

```python

grouped = data.groupby('listed_in(type)')['votes'].sum()


result = pd.DataFrame({'votes': grouped})


plt.bar(result.index, result['votes'], color="green")

plt.xlabel("Type of Restaurant", color="red", size=35)
plt.ylabel("Votes", color="red", size=20)


plt.xticks(rotation=45, ha="right")


plt.show()
```

# Task.3. What are the ratings that the majority of restaurants have received?

```python
plt.hist(data['rate'],bins=5)
plt.title("rest_type")
plt.show()
```

# Task.4. Which mode (online or offline) has received the maximum rating?

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=data)
```

# Task.5. Which type of restaurant received more offline orders, so that Zomato can provide those customers with some good offers?

```python
pivot_table=data.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
